```python
import numpy as np

M = 20
f = 20
Msol = 4.925491025543575903411922162094833998e-6
Omega_0 = M*Msol*np.pi*f

q = [2,1.5,2]
qs = ["20","15","20"]
chi1 = [.9,0.02,0.1]
chi2 = [.5,0.01,-0.1]
chi1s = ["9","02","1"]
chi2s = ["5","01","m1"]
pert = 1. + 4e-14

k = 2
pyseobnr_dynamics_label = f"./pyseobnr_dynamics_q_{qs[k]}_chi1_{chi1s[k]}_chi2_{chi2s[k]}.dat"
pyseobnrpert_dynamics_label = f"./pyseobnr_pertO14_dynamics_q_{qs[k]}_chi1_{chi1s[k]}_chi2_{chi2s[k]}.dat"
our_dynamics_label = f"./our_dynamics_q_{qs[k]}_chi1_{chi1s[k]}_chi2_{chi2s[k]}.dat"
#nocalib_dynamics_label = f"./nocalib_dynamics_q_{qs[k]}_chi1_{chi1s[k]}_chi2_{chi2s[k]}.dat"
err_trustedpert_label = f"./err_trusted_pert_q_{qs[k]}_chi1_{chi1s[k]}_chi2_{chi2s[k]}.dat"
err_trustedours_label = f"./err_trusted_ours_q_{qs[k]}_chi1_{chi1s[k]}_chi2_{chi2s[k]}.dat"
#err_trustednocalib_label = f"./err_trusted_nocalib_q_{qs[k]}_chi1_{chi1s[k]}_chi2_{chi2s[k]}.dat"
    

trusted_dynamics = np.loadtxt(pyseobnr_dynamics_label)
pert_dynamics = np.loadtxt(pyseobnrpert_dynamics_label)
our_dynamics = np.loadtxt(our_dynamics_label)
#nocalib_dynamics = np.loadtxt(nocalib_dynamics_label)
err_trusted_pert = np.loadtxt(err_trustedpert_label)
err_trusted_ours = np.loadtxt(err_trustedours_label)
#err_trusted_nocalib = np.loadtxt(err_trustednocalib_label)
labels = [r'$r$',r'$\phi$',r'$p_{r_{*}}$',r'$p_{\phi}$']
savefig_labels = ['radius','orbital_phase','tortoise_momentum','angular_momentum']
```

```python
import matplotlib.pyplot as plt
import numpy as np

idx_to_plot = 1

# Create subplots with 2 rows and 1 column
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Plot on the first subplot
ax1.plot(trusted_dynamics[:,0], trusted_dynamics[:,idx_to_plot],color = 'black',label = 'pyseobnr')
ax1.scatter(our_dynamics[:,0], our_dynamics[:,idx_to_plot],s = (1/72)**(-0.25) , label = 'ours')
#ax1.scatter(nocalib_dynamics[:,0], nocalib_dynamics[:,idx_to_plot],s = (1/72)**(-0.25) , label = 'nocalib')
ax1.set_ylabel(labels[idx_to_plot-1])
ax1.legend()

# Plot on the second subplot (log scale for y-axis)
ax2.plot(err_trusted_pert[:,0], err_trusted_pert[:,idx_to_plot],label = 'trusted-pert')
ax2.plot(err_trusted_pert[:,0], 3*err_trusted_pert[:,idx_to_plot],label = '3x trusted-pert',color = 'black',linestyle = 'dashed')
ax2.plot(err_trusted_ours[:,0], err_trusted_ours[:,idx_to_plot],label = 'trusted-ours')
#ax2.plot(err_trusted_nocalib[:,0], err_trusted_nocalib[:,idx_to_plot],label = 'trusted-nocalib')
ax2.set_yscale('log')
ax2.set_xlabel(r'$t$')
ax2.set_ylabel(r'$E_{\rm rel}($ '+labels[idx_to_plot-1]+r' $)$')
ax2.legend()

# Remove the space between subplots
plt.subplots_adjust(hspace=0)
plot_label = f"{savefig_labels[idx_to_plot-1]}_comparison_q_{qs[k]}_chi1_{chi1s[k]}_chi2_{chi2s[k]}.png"
plt.savefig(plot_label,dpi = 300)
# Show the plot
plt.show()
```
