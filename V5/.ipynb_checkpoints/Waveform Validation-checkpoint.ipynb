```python
import numpy as np

M = 20
f = 20
Msol = 4.925491025543575903411922162094833998e-6
Omega_0 = M*Msol*np.pi*f
q = 1.5
qs = "15"
chi1 = 0.02
chi2 = 0.01
chi1s = "02"
chi2s = "01"

m1 = q/(1. + q)
m2 = 1./(1. + q)
ap = m1*chi1 + m2*chi2
am = m1*chi1 - m2*chi2
nu = m1*m2

h22_BOB_calib = np.loadtxt(f"h22_v5HMcalib_BOB_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat")
h22_BOB_NQC = np.loadtxt(f"h22_v5HMcalib_BOB_inspiral_NQC_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat",dtype = np.cdouble)
h22_BOB_noNQC = np.loadtxt(f"h22_v5HMcalib_BOB_inspiral_no_NQC_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat",dtype = np.cdouble)
dynamics_BOB = np.loadtxt(f"v5HMcalib_BOB_inspiral_dynamics_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat")

h22_BOBcalibDeltat_calib = np.loadtxt(f"h22_v5HMcaliba6dSO_BOBcalibDeltat_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat")
h22_BOBcalibDeltat_NQC = np.loadtxt(f"h22_v5HMcaliba6dSO_BOBcalibDeltat_inspiral_NQC_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat",dtype = np.cdouble)
h22_BOBcalibDeltat_noNQC = np.loadtxt(f"h22_v5HMcaliba6dSO_BOBcalibDeltat_inspiral_no_NQC_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat",dtype = np.cdouble)
dynamics_BOBcalibDeltat = np.loadtxt(f"v5HMcaliba6dSO_BOBcalibDeltat_inspiral_dynamics_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat")


h22_v5HM = np.loadtxt(f"h22_v5HM_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat")
dynamics_v5HM = np.loadtxt(f"v5HM_inspiral_dynamics_q_{qs}_chi1_{chi1s}_chi2_{chi2s}.dat")
```

```python
import matplotlib.pyplot as plt

# first plot dynamics

idx_to_plot = 1

plt.plot(dynamics_v5HM[:,0], dynamics_v5HM[:,idx_to_plot],color = 'black',label = 'pyseobnr')
plt.scatter(dynamics_BOB[:,0], dynamics_BOB[:,idx_to_plot],s = (1/72)**(-0.25) , label = 'ours(v5HM calib)')
plt.scatter(dynamics_BOBcalibDeltat[:,0], dynamics_BOBcalibDeltat[:,idx_to_plot],s = (1/72)**(-0.25) , label = 'ours(v5HM calib)')
plt.legend()
# Show the plot
plt.show()
```

```python
# plot the non-nqc-ed waveform
dT = 2.4627455127717882e-05/M/4.925490947641266978197229498498379006e-6
t_new = np.arange(dynamics_BOB[0,0],dynamics_BOB[-1,0],dT)
t_new_BOB = np.arange(dynamics_BOBcalibDeltat[0,0],dynamics_BOBcalibDeltat[-1,0],dT)
plt.plot(h22_v5HM[:,0], h22_v5HM[:,1],color = 'black',label = 'pyseobnr')
plt.plot(t_new - t_new[0] + h22_v5HM[0,0],np.real(h22_BOB_noNQC),label = 'ours, no NQC')
#plt.plot(t_new_BOB - t_new_BOB[0] + h22_v5HM[0,0],np.real(h22_BOBcalibDeltat_noNQC),label = 'ours, BOB Deltat, no NQC')
plt.xlim(-100,50)
plt.legend()
plt.show()
```

```python
# plot the nqc-ed waveform
dT = 2.4627455127717882e-05/M/4.925490947641266978197229498498379006e-6
t_new = np.arange(dynamics_BOB[0,0],dynamics_BOB[-1,0],dT)
t_new_BOB = np.arange(dynamics_BOBcalibDeltat[0,0],dynamics_BOBcalibDeltat[-1,0],dT)
plt.plot(h22_v5HM[:,0], h22_v5HM[:,1],color = 'black',label = 'pyseobnr')
plt.plot(t_new - t_new[0] + h22_v5HM[0,0],np.real(h22_BOB_NQC),label = 'ours, NQC')
#plt.plot(t_new_BOB - t_new_BOB[0] + h22_v5HM[0,0],np.real(h22_BOBcalibDeltat_NQC),label = 'ours, BOB Deltat, NQC')
plt.xlim(- 100,50)
plt.ylim(-0.4,0.4)
plt.legend()
plt.show()
```

```python
# plot the full waveform amplitude
plt.plot(h22_v5HM[:,0], np.abs(h22_v5HM[:,1] + 1j*h22_v5HM[:,2]),color = 'black',label = 'pyseobnr')
plt.plot(h22_BOB_calib[:,0],np.abs(h22_BOB_calib[:,1] + 1j*h22_BOB_calib[:,2]),label = 'SEOBNRv5HM + BOB')
#plt.plot(h22_BOBcalibDeltat_calib[:,0],np.abs(h22_BOBcalibDeltat_calib[:,1] + 1j*h22_BOBcalibDeltat_calib[:,2]),label = 'ours, BOB Deltat')
plt.xlim(-100,50)
plt.xlabel(r'$t(M)$')
plt.ylabel(r'$h_{+}$')
plt.legend()
plt.savefig(f"hamp_comparison_q_{qs}_chi1_{chi1s}_chi2_{chi2s}")
plt.show()
```

```python
# plot the full waveform phase

plt.plot(h22_v5HM[:,0], np.unwrap(np.angle(h22_v5HM[:,1]+1j*h22_v5HM[:,2])),color = 'black',label = 'pyseobnr')
plt.plot(h22_BOB_calib[:,0],np.unwrap(np.angle(h22_BOB_calib[:,1]+1j*h22_BOB_calib[:,2])),label = 'ours')
#plt.plot(h22_BOBcalibDeltat_calib[:,0],np.unwrap(np.angle(h22_BOBcalibDeltat_calib[:,1]+1j*h22_BOBcalibDeltat_calib[:,2])),label = 'ours')
plt.xlim(-100,50)
plt.ylim(1100,1300)
plt.legend()
plt.show()
```

```python
# plot h_+ (match phases at peak)
v5HM_amp = np.abs(h22_v5HM[:,1] - 1j*h22_v5HM[:,2])    
v5HM_phase = np.unwrap(np.angle(h22_v5HM[:,1] - 1j*h22_v5HM[:,2]))

BOB_amp = np.abs(h22_BOB_calib[:,1] - 1j*h22_BOB_calib[:,2])
BOB_phase = np.unwrap(np.angle(h22_BOB_calib[:,1] - 1j*h22_BOB_calib[:,2]))
BOB_idx_phase_match = np.argmin(np.abs(h22_BOB_calib[:,0]))

BOBcalibDeltat_amp = np.abs(h22_BOBcalibDeltat_calib[:,1] - 1j*h22_BOBcalibDeltat_calib[:,2])
BOBcalibDeltat_phase = np.unwrap(np.angle(h22_BOBcalibDeltat_calib[:,1] - 1j*h22_BOBcalibDeltat_calib[:,2]))
BOBcalibDeltat_idx_phase_match = np.argmin(np.abs(h22_BOBcalibDeltat_calib[:,0]))


BOB_v5HM_phase_match = np.interp(h22_BOB_calib[:,0],h22_v5HM[:,0],v5HM_phase)[BOB_idx_phase_match]
BOB_phase = BOB_phase - BOB_phase[BOB_idx_phase_match] + BOB_v5HM_phase_match

BOBcalibDeltat_v5HM_phase_match = np.interp(h22_BOBcalibDeltat_calib[:,0],h22_v5HM[:,0],v5HM_phase)[BOBcalibDeltat_idx_phase_match]
BOBcalibDeltat_phase = BOBcalibDeltat_phase - BOBcalibDeltat_phase[BOBcalibDeltat_idx_phase_match] + BOBcalibDeltat_v5HM_phase_match

h22_BOB_plus = BOB_amp*np.cos(BOB_phase)
h22_BOB_cross = BOB_amp*np.sin(-1*BOB_phase)

h22_BOBcalibDeltat_plus = BOBcalibDeltat_amp*np.cos(BOBcalibDeltat_phase)
h22_BOBcalibDeltat_cross = BOBcalibDeltat_amp*np.sin(-1*BOBcalibDeltat_phase)

plt.plot(h22_v5HM[:,0],h22_v5HM[:,1],color = 'black',label = 'pyseobnr')
plt.plot(h22_BOB_calib[:,0],h22_BOB_plus,label = 'SEOBNRv5HM + BOB')
#plt.plot(h22_BOBcalibDeltat_calib[:,0],h22_BOBcalibDeltat_plus,label = 'ours, BOB Deltat')
plt.xlim(-100,50)
plt.xlabel(r'$t(M)$')
plt.ylabel(r'$h_{+}$')
plt.legend()
plt.savefig(f"hplus_comparison_q_{qs}_chi1_{chi1s}_chi2_{chi2s}")
plt.show()
```

```python
# plot h_x (match phases at peak)

plt.plot(h22_v5HM[:,0],h22_v5HM[:,2],color = 'black',label = 'pyseobnr')
plt.plot(h22_BOB_calib[:,0],h22_BOB_cross,label = 'SEOBNRv5HM + BOB')
#plt.plot(h22_BOBcalibDeltat_calib[:,0],h22_BOBcalibDeltat_cross,label = 'ours, BOB Deltat')
plt.xlim(-100,50)
plt.legend()
plt.xlabel(r'$t(M)$')
plt.ylabel(r'$h_{\times}$')
plt.savefig(f"hcross_comparison_q_{qs}_chi1_{chi1s}_chi2_{chi2s}")
plt.show()
```

```python
# Amplitude derivatives
BOB_damp = np.gradient(BOB_amp, h22_BOB_calib[:,0])
v5HM_damp = np.gradient(v5HM_amp,h22_v5HM[:,0])
plt.plot(h22_v5HM[:,0],v5HM_damp)
plt.plot(h22_BOB_calib[:,0],BOB_damp)
plt.xlim(-100,50)
```

```python
BOB_ddamp = np.gradient(BOB_damp, h22_BOB_calib[:,0])
v5HM_ddamp = np.gradient(v5HM_damp,h22_v5HM[:,0])
plt.plot(h22_v5HM[:,0],v5HM_ddamp)
plt.plot(h22_BOB_calib[:,0],BOB_ddamp)
plt.xlim(-100,50)
```

```python
# Phase derivatives

BOB_dphi = np.gradient(np.abs(BOB_phase), h22_BOB_calib[:,0])
v5HM_dphi = np.gradient(np.abs(v5HM_phase),h22_v5HM[:,0])
plt.plot(h22_v5HM[:,0],v5HM_dphi)
plt.plot(h22_BOB_calib[:,0],BOB_dphi)
plt.xlim(-100,50)
```

```python
BOB_ddphi = np.gradient(BOB_dphi, h22_BOB_calib[:,0])
v5HM_ddphi = np.gradient(v5HM_dphi,h22_v5HM[:,0])
plt.plot(h22_v5HM[:,0],v5HM_ddphi)
plt.plot(h22_BOB_calib[:,0],BOB_ddphi)
plt.xlim(-50,50)
plt.ylim(0,0.012)
```
