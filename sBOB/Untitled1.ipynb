```python
import numpy as np
import matplotlib.pyplot as plt

dat = np.loadtxt('sBOB_waveform.txt')
print(dat)
```

```python
plt.plot(dat[:,0],np.abs(dat[:,3])*np.cos(dat[:,4]), label = r'$h_+$')
plt.plot(dat[:,0],np.abs(dat[:,3])*np.sin(dat[:,4]), label = r'$h_x$')
plt.xlim(-50,50)
plt.xlabel('time($M$)')
plt.legend()
plt.savefig('sBOB.png',dpi = 300)
```
